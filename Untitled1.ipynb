```python
import numpy as np
import pandas as pd
from sklearn import datasets
from scipy.stats import chi2_contingency, chi2
import math
```

```python
def load_iris():

    iris = datasets.load_iris()#Hint: use datasets from sklearn
    return iris
```

```python
def represent_data(iris):
  
    #The column names of the new table
    col_names = ['sepal length', 'sepal width', 'petal length', 'petal width', 'species']
    #Constructing the new table
    data = pd.DataFrame(iris.data)#Hint: use pandas to construct the new data format with col_names as the columns names
    data.rename(columns = {0:col_names[0], 1:col_names[1], 2:col_names[2], 3:col_names[3]}, inplace=True)
    name = iris.target
    data[col_names[4]] = name
    print('Samples from the Data:')
    display(data.head())
    return data
```

```python
def calculate_chi2_value(freq_table, expected_table):
  
    freq = pd.DataFrame(freq_table).to_numpy()
    expected = pd.DataFrame(expected_table).to_numpy()
    col = len(freq_table.columns) 
    row = len(freq_table.axes[0]) 
    chi2_value = 0
    for i in range(row ):
      for j in range(col ):
         chi2_value = chi2_value + (math.pow(freq[i][j] - expected[i][j],2))/ expected[i][j]
    return chi2_value
```

```python
def calculate_p_value(chi2_value, df):
  
    p_value = 1 - chi2.cdf(chi2_value,df)

    return p_value
```

```python
def tabulate(data, label1, label2):
    
    freq_table = pd.crosstab(index=data[label1], columns=data[label2], margins = True)    
    freq = pd.DataFrame(freq_table).to_numpy() #convert frequency dataframe into array   
    col = len(freq_table.columns) #number of rows
    row = len(freq_table.axes[0]) #number of columns
    df = (col-2) * (row - 2)
    exp = [[0]*(col -1)]* (row-1) #initialize exp array with value 0
    for i in range(row - 1) : 
      for j in range(col - 1) :
        exp[i][j] = (freq[row - 1][j] * freq[i][col - 1]) / freq[row - 1][col - 1] #expected frequency calculation
    cols = []
    for i in range(col-1):
      cols.append(i) 
    expected_table = pd.DataFrame(exp, columns = cols)#col names of exp table
    freq_table = pd.crosstab(index=data[label1], columns=data[label2])
    chi2_value = calculate_chi2_value(freq_table, expected_table)
    p_value = calculate_p_value(chi2_value, df) 
    print('The observed tabel:')
    display(freq_table)
   
    return freq_table, chi2_value, p_value, df, expected_table
```

```python
def assert_results(freq_table, chi2_value, p_value, df, expected_table):
   
    good = True
    
    stat, p, dof, expected = chi2_contingency(freq_table)
    if(round(chi2_value,5) != round(stat,5)) :
      good = False
      print(stat)
    if(round(p_value,5) != round(p,5)):
      good = False
      print(p)
    if(df != dof):
      good = False
      print(dof)
    exp = pd.DataFrame(expected_table).to_numpy()
    col = len(freq_table.columns) 
    row = len(freq_table.axes[0])
    for i in range(row - 1):
      for j in range(col - 1):
        if(expected[i][j] != exp[i][j]):
          good = False
          break
        if (good == False):
          break 
    
    return good
```

```python

#Loading the dataset
iris = load_iris()

#Tabulating the raw data
data = represent_data(iris)

#Define the level of significance
alpha = 0.1

col_names = data.columns
n = len(col_names)
#Traverse the features except the target label
for j in range(0, n-1):
    print('Working on the target label(species) vs. ' + col_names[j])
    
    #Apply the independence test
    freq_table, chi2_value, p_value, df, expected_table = tabulate(data, 'species', col_names[j])
    print('\t-chi square sum: ' + str(chi2_value))
    print('\t-p-value: ' + str(p_value))
    print('\t-degrees of freedom: ' + str(df))
    
    #Check the results
    good = assert_results(freq_table, chi2_value, p_value, df, expected_table)
    
    print('\t-Are these results correct? --- ' + str(good))
    if p_value < alpha:
        print('\t-The target label is dependent on the feature: ' + col_names[j])
    else:
        print('\t-The target label: is independent of the label: ' + col_names[j])
    print('\n')
```
